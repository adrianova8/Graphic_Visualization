# **Graphic Display of a Dataset**

**TASK INFORMATION:** Complements data exploration techniques through graphical visualization, with the Matplotlib and Searborn libraries.

## **DATASET**

```python
#First at all, we import everything we need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#We load our dataset about flights
airline_df = pd.read_csv('new_DelayedFlights.csv')

# Removing the first column "Unnamed:0"
airline_df = airline_df.drop(["Unnamed: 0"], axis = 1) 
airline_df.head()
```

## **Exercise 1**

### **1.1** Summarize the Delayed Flights.csv dataset graphically

####  **1.1.1** Flight by Carrier

* The **BEST** 3 carriers in terms of flights:
    + WN Southwest Airlines
    + AA American Airlines
    + MQ Envoy Air
    
    
* The **WORST** 3 carriers in terms of flights:
    + AQ 9 Air
    + HA Hawaiian Airlines
    + F9 Frontier Airlines

```python
# Flights by carrier

labels_uniquecarrier = airline_df['uniquecarrier'].unique().tolist()
flight_by_uniquecarrier = airline_df.uniquecarrier.value_counts()


#Important keep the index
x = np.arange(len(labels_uniquecarrier))  # the label locations
width = 0.5 # the width of the bars

fig, ax = plt.subplots()
plot1 = ax.bar(x-width/2, flight_by_uniquecarrier, width, label='Unique Carrier', color='#7FFFD4',alpha=0.55)

# We add some text for labels, title and custom x-axis tick labels.
ax.set_xlabel('Unique Carrier')
ax.set_ylabel('Nº of flights')
ax.set_title('Flights by Unique Carrier (USA, 2008)')
ax.set_xticks(x) # We add ALL the labels of Uniquecarrier (number)
ax.set_xticklabels(flight_by_uniquecarrier.index) # We add the name of the labels
ax.legend()

fig.set_size_inches(15, 5) #We add extra inches to see it better
ax.bar_label(plot1, padding=3) # We add padding on the bar labels
fig.tight_layout() # The layout is tight

plt.savefig('figure1.png') # We save the plot in our computer as png.

plt.show()
```

```python
# Flights by Carrier (seaborns)

g = sns.displot(data=airline_df, x='uniquecarrier', hue='uniquecarrier', palette='Set2')
g.fig.set_size_inches(15,5)
g.ax.set_title('Nº of Flights per Carrier (USA, 2008)')

g.savefig('figure2.png')

plt.show()
```

#### **1.1.2** Monthly flights

* The **BEST** 3 months of the year (flights):
    + December
    + June
    + March

* The **WORST** 3 months of the year (flights):
    + September
    + October
    + November

```python
# Monthly flights

#Fixing random state for reproducibility
np.random.seed(48265148)

#Data
label_months = airline_df.month.value_counts()
label_months = label_months.sort_index()
label_months.index = ['January', 'February', 'March', 'April', 'May', 'June',
                         'July','August', 'September','October', 'November','December']
y_pos = np.arange(len(label_months))

fig, ax = plt.subplots()

hbars = ax.barh(y_pos, label_months, color='#FFC0CB',alpha=0.55)
ax.set_yticks(y_pos)
ax.set_yticklabels(label_months.index)
ax.set_ylabel('Months of the Year')
ax.set_xlabel('Nº of Flights')
ax.set_title('Monthly Flights (USA,2008)')

# Label with specially formatted floats
ax.bar_label(hbars, fmt = '%.0f')
fig.set_size_inches(14, 5)

plt.show()
```

```python
# Monthly Flights (seaborns)

g = sns.barplot(data = airline_df, x = label_months, y = label_months.index, palette='dark:salmon_r')
g.set_xlabel('Nº of Flights')
g.set_ylabel('Months')
g.set_title('Monthly Flights (USA, 2008)')
g.figure.set_figheight(5)
g.figure.set_figwidth(12)

plt.savefig('figure3.png')

plt.show()
```

#### **1.1.3** Accumulated Delay 

* The range of the delay of the airlines goes from one hour of the expected (-60 minutes) to 4 hours of delay (240 minutes). The intervals for each bar of the histogram is 15 minutes.


* We can see how most airlines arrive on time or a few minutes late (max 50 minutes). There are some airlines that arrive earlier than expected and others arrive too late (>50 minutes), but this tend goes down with the past of the minutes.

```python
# Acumulated Delay. A histogram to show the frequency of Arrival Delays.

arrival_delay = airline_df.arrdelay # Our variable of interest

fig = plt.figure(figsize = (10,7))
plt.hist(arrival_delay,bins = 20, range = (-60,240), color = '#C0C0C0', alpha = 0.55) 

# Details
plt.xlabel('MINUTES DELAY AT ARRIVAL')
plt.ylabel('FREQUENCY (Nº OBSERVATIONS)')
plt.title('HISTOGRAM OF ARRIVAL DELAYS (USA,2008)')
plt.grid(True)

plt.savefig('figure4.png')

plt.show()
```

```python
#The same example, but introducing a 'best fit' line

mu = 15  # mean of distribution
sigma = 20  # standard deviation of distribution
arrival_delay = airline_df.arrdelay

fig, ax = plt.subplots(figsize=(8,5))

# the histogram
n, bins, patches = ax.hist(arrival_delay, bins= 20, range=(-60,240), density=True, color='#ADD8E6', alpha = 0.55)

# 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('MINUTES DELAY AT ARRIVAL')
ax.set_ylabel('FREQUENCY (Nº OBSERVATIONS)')
ax.set_title('HISTOGRAM OF ARRIVAL DELAYS (USA, 2008)')

plt.grid(True)
plt.show()
```

#### **1.1.4** Delays by Airline (mean)

* For this section we have done the mean for each airline to know which has most mean delays. 


 + The best 3 companies in terms of delays are:
     - AQ: 9 Air Co Ltd
     - F9: Frontier Airlines
     - WN: Southwest Airlines
 + The worst 3 companies in terms of delays are:
     - YV: Mesa Airlines
     - B6: Jetblue Airways
     - OH: PSA Airlines
     

```python
# Delays by Airline

arrdelay_by_carrier = airline_df.arrdelay.groupby(airline_df.uniquecarrier) # We create a series of arrdelay by carrier
arrdelay_by_carrier = arrdelay_by_carrier.mean() # We relativize the values making the mean
arrdelay_by_carrier = arrdelay_by_carrier.sort_values(ascending = False) # Sort the carriers by maximum value

# The plot
fig, ax = plt.subplots(figsize = (10,7))
plot = ax.bar(arrdelay_by_carrier.index, arrdelay_by_carrier, width=0.5, color='#E6E6FA', alpha=0.55) 

ax.set_ylabel('Mean Arrival Delays (minutes)')
ax.set_xlabel('Unique Carrier (airlines)')
ax.set_title('Mean Arrival Delay by Unique Carrier (USA, 2008)')
ax.bar_label(plot, fmt = '%.1f')
fig.set_size_inches(14, 5)

plt.show()


```

#### **1.1.5** Relation between Departure and Arrival  Delay of Airlines

+ Most flights that depart late end up arriving late at the same rate. At the beginning of the graph we observe a base that some flights leave X minutes late and arrive at their destination even later (some minutes).

+ We also found some outliers. Flights leaving X minutes late arriving at their destination much later than expected. 

```python
# Relation departure and arrival delay

arrival_delay = airline_df.arrdelay
departure_delay = airline_df.depdelay

fig = plt.figure(figsize = (10,7))
plt.scatter(arrival_delay, departure_delay, alpha = 0.3, color = '#FBDD7E')
plt.xlim(-60, 1200) # Limit the range of arrival to 60 minutes before expected and 20 hours delayed
plt.ylim(-60, 1200) # Limit the range of departure to 60 minutes before expected and 20 hours delayed
plt.xlabel('Delayed at Arrival (minutes)')
plt.ylabel('Delayed at Departure (minutes)')
plt.title(' Flight Delays (USA, 2008)')

plt.grid()
plt.show()
```

```python
# Relation departure and arrival delay (seaborns)

g = sns.relplot(data = airline_df, x = 'arrdelay', y = 'depdelay', hue = 'uniquecarrier')
g.fig.set_size_inches(10,4)
g.ax.set_title('Flight Delays (USA, 2008)')

g.savefig('figure5.png')

plt.show()
```

#### **1.1.6** Relation between Departure and Arrival Delay by Carrier

* The trend is that those flights with the longest departure delay have the same arrival delay.


* Some of the flights, despite leaving late, manage to cut minutes on arrival, such as AQ, WN or AS carriers. On the contrary, some flights take longer to arrive than to depart, on average, such as, OH, FL or NW.

```python
# Relation departure and arrival delay by carrier

arrival_delay = airline_df.arrdelay.groupby(airline_df.uniquecarrier).mean()
departure_delay = airline_df.depdelay.groupby(airline_df.uniquecarrier).mean()
label_carrier = airline_df.uniquecarrier.unique()
label_carrier_tidy = label_carrier.sort()

x = np.arange(len(label_carrier))
width = 0.50

fig, ax = plt.subplots(figsize = (15,8))
plots1 = ax.bar(x-width/2, departure_delay, width, label = 'Departure Delay', color ='#E6E6FA')
plots2 = ax.bar(x+width/2, arrival_delay, width, label = 'Arrival Delay', color = '#C79FEF')

ax.set_ylabel('Mean Delay (minutes)')
ax.set_xlabel('Unique Carrier')
ax.set_title('Mean Departure and Arrival Delay by Carrier (USA,2008)')
ax.set_xticks(x)
ax.set_xticklabels(label_carrier)
ax.bar_label(plots1, fmt = '%.1f')
ax.bar_label(plots2, fmt = '%.1f')
ax.legend()


fig.tight_layout()

plt.show()
```

```python
# Second Method (more productive)

# First at all we have to divide by 'groupy' our variables of interest
carrier_delays = airline_df[['depdelay', 'arrdelay']].groupby(airline_df.uniquecarrier).mean()
carrier_delays = carrier_delays.sort_values('arrdelay')

# We request what we will need in the exercise
label_carrier = carrier_delays.index
departure_delay_mean = carrier_delays.depdelay
arrival_delay_mean = carrier_delays.arrdelay

x = np.arange(len(label_carrier))
width = 0.5

fig, ax = plt.subplots(figsize = (15,8))
plot1 = ax.bar( x-width/2, departure_delay_mean, width, label='Departure Delay', color = '#D1B26F')
plot2 = ax.bar( x+width/2, arrival_delay_mean, width, label='Arrival Delay', color = '#FBDD7E')

ax.set_ylabel('Mean Delay (minutes)')
ax.set_xlabel('Unique Carrier')
ax.set_title('Mean Departure and Arrival Delay by Carrier (USA,2008)')
ax.set_xticks(x)
ax.set_xticklabels(label_carrier)
ax.bar_label(plot1, fmt = '%.1f')
ax.bar_label(plot2, fmt = '%.1f')
ax.legend()

fig.tight_layout()
plt.savefig('figure6.png')

plt.show()
```

#### **1.1.7** Relation between Departure and Arrival Delay, Air Time and Unique Carrier

* We can say some relevant conclusions from this graph:

     + There are some flights of short duration that also have very high delays, on average. This fact makes people think twice about traveling with this type of carrier. Exemple: MQ, OO, 9E, EV or YV.
     
     + On the contrary, there are other carriers that despite having delays in their flights, if their trajectory is greater, people do not take delays into account as much. Example: CO, AA, UA or B6.


```python
#  Relation between Departure and Arrival Delay, Air Time and Unique Carrier

carrier_delays_and_airtime = airline_df[['depdelay','arrdelay','airtime']].groupby(airline_df.uniquecarrier).mean()
carrier_delays_and_airtime = carrier_delays_and_airtime.sort_values('arrdelay')

#Request what we will use
labels_carrier = carrier_delays_and_airtime.index
departure_delay_mean = carrier_delays_and_airtime.depdelay
arrival_delay_mean = carrier_delays_and_airtime.arrdelay
airtime_mean = carrier_delays_and_airtime.airtime

x = np.arange(len(labels_carrier))  
width = 0.9 

fig, ax = plt.subplots(figsize=(11,5))
plot1 = ax.bar(x - width/3, departure_delay_mean, width/3, label='Departure Delay', color='#EE82EE')
plot2 = ax.bar(x , arrival_delay_mean, width/3, label='Arrival Delay',color='#800080')
plot3 = ax.bar(x + width/3, airtime_mean, width/3, label='Air Time', color='#C5C9C7')

ax.set_ylabel('Minutes')
ax.set_xlabel('Unique Carrier')
ax.set_title('Mean Arrival & Departure Delay and Mean Air Time by Carrier (USA, 2008)')
ax.set_xticks(x)
ax.set_xticklabels(labels_carrier)
ax.legend()

fig.tight_layout()

plt.savefig('figure7.png')

plt.show()
```
