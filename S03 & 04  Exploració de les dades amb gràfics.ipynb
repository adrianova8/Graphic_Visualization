### **Create new columns (average flight speed, whether late or not ...)**

#### **Average flight speed**

```python
#We are going to calculate the average flight speed and add it into our data set

airline_df["average_flight_speed(km/h)"] = ((airline_df["distance"] * 1.60934) / (airline_df["airtime"] / 60)).round(2)
airline_df.head()
```

```python
# Flight Speed Chart

#Check which rows have no AirTime recorded.
airline_df[airline_df.airtime == 0]

#Replace infinity values in data set with NaN
airline_df.replace([np.inf, -np.inf], np.nan, inplace = True)

#Now, we can create the flight speed chart's histogram 

flight_speed = airline_df['average_flight_speed(km/h)']

fig = plt.figure(figsize=(10,7))
plt.hist(flight_speed, bins = 20, range = (200,1000), color = '#FF7F50') # #20 histogram bins in intervals of 40km/h between 200 and 1000
plt.xlabel('Average Flight Speed (KM/H)')
plt.ylabel('Frequency')
plt.title('Histogram of Avg Flight Speed (USA, 2008)')

plt.show()
```

* Most of the flights in United States (2008), its average speed was [550-750 km/h]

#### **Flight was late or not**

```python
#We are going to construct a dichotomous variable. 1 if is late 0 on the contrary
#We add the new variable in our data set

def delayedFligt(x):
    if x > 0:
        return 1
    else:
        return 0

airline_df["delayed_flight"] = (airline_df["arrdelay"].apply(delayedFligt))
airline_df.head()
```

```python
# Flight was Late or Not Chart

x = airline_df.delayed_flight.groupby(airline_df.delayed_flight).count()
xs = pd.DataFrame(x)

labels = x.index
colors = ['#FFC0CB', '#FF81C0']

fig = plt.figure(figsize=(7,5))
plt.pie(x, labels = labels, startangle = 120, autopct='%0.2f %%', colors = colors)
plt.legend(title='Delayed Flights:', loc='right', bbox_to_anchor=(1.5, 0.7))
plt.suptitle('Percentage of Late Flights (USA, 2008)')

plt.show()
```

### **Table of airlines with the most accumulated arrears**

* We want to know the relative delayed flights. For that reason we distinguish the airline carriers per the total quantity of flights and his relative delays.

```python
# First, we want to know the number of flights per airline carrier and his %
airlines_flights = airline_df.uniquecarrier.value_counts()
airlines_percent = airline_df.uniquecarrier.value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
carriers_df = pd.DataFrame({'annual flights': airlines_flights, 'percent of flights': airlines_percent})

# Second, we want to know the relative delayed flights per airline carrier

grouped = airline_df.groupby('uniquecarrier')
carrier_delays = (grouped['delayed_flight'].agg(np.sum))
carriers_df['annual delays'] = carrier_delays
carriers_df['percent of flights delayed'] = ((carriers_df['annual delays']/(carriers_df['annual flights']))*100).round(2).astype(str) + '%'
carrier_delays = carriers_df.sort_values(by=['percent of flights delayed'],ascending=False).copy()
carrier_delays
```

**THE TOP 3 PERFORMERS FOR ARRIVAL DELAYS:**
    
    - CO -> North-Western Cargo International Airlines
    - US -> Puerto Rico International Airlines
    - WN -> Southwest Airlines
    
**THE WORST 3 PERFORMERS FOR ARRIVAL DELAYS:**

    - HA -> Hawaiian Airlines
    - YV -> Mesa Airlines
    - OH -> PSA Airlines

```python
# Annual Delays on Total Flights (relative)

carriers_df['delayed_flights_by_carrier'] = ((carriers_df['annual delays']/(carriers_df['annual flights']))) #Create the variable without the %
carrier_delays = carriers_df.sort_values(by = ['delayed_flights_by_carrier'],ascending=False).copy()
x = carrier_delays.delayed_flights_by_carrier

x_pos = np.arange(len(x.index))

fig = plt.figure(figsize=(15,5))
plt.bar(x_pos, x, color = '#F0FFFF', edgecolor='blue')
plt.xticks(x_pos, x.index)
plt.ylabel('Delays/Total Flights')
plt.xlabel('Carriers')
plt.title('Delayed Flights (USA, 2008)')

plt.show()
```

### **What are the longest flights? And the most backward?**

#### **Longest Flights**

```python
#DISTANCE (MILES)

grouped = airline_df.groupby(['uniquecarrier','flightnum','origin','dest']) #Select the interesting columns
longest_flights = (grouped['distance'].agg(np.mean)) #add 'distance' to the group and make the mean
longest_flights = longest_flights.sort_values(ascending=False) #reordering
longest_flights.head(10)

```

```python
# Longest Flights Chart

longest_flights = airline_df.distance.astype(str) + "-" + airline_df.origin + "-" + airline_df.dest

#Create a Data Frame with the values of Distance - Origin - Destination
longest_flights = longest_flights.unique()
longest_flights = pd.DataFrame(longest_flights)

# Separate the variables in 3 columns
longest_flights = longest_flights[0].str.split('-', expand=True)

#String number to a pure number
longest_flights[0] = longest_flights[0].astype(int)

# Sort the data and only the first 10 numbers
longest_flights = longest_flights.sort_values(by = 0, ascending=False).reset_index(drop=True).head(10)

# The Chart
fig = plt.figure(figsize=(15,5))
g = sns.barplot( x = longest_flights[1] +'-' + longest_flights[2], y = longest_flights[0])
g.set_xlabel('Origin - Destination')
g.set_ylabel('Distance (KM)')
g.set_title('Longest Flights (USA, 2008)')

plt.show()
```

#### **Most delayed flights**

```python
# The bigger 'arrdelay' is, the bigger the delay

grouped = airline_df.groupby(['uniquecarrier','flightnum','origin','dest'])
delayed_flights = (grouped['arrdelay'].agg(np.mean))
delayed_flights = delayed_flights.sort_values(ascending=False)
delayed_flights.head(5)
```

```python
# Delayed Flights by origin-destination

#Create a variable of ORIGIN-DEST
airline_df['Origin-Dest'] = airline_df.origin + '-' + airline_df.dest

#Select the variable of arrdelay to do the mean
flights = airline_df.arrdelay.groupby(airline_df['Origin-Dest']).mean()
flights = flights.sort_values(ascending=False).head(10)


# The Chart
fig = plt.figure(figsize=(15,5))
g = sns.barplot(x = flights.index, y = flights)
g.set_xlabel('Origin - Destination')
g.set_ylabel('Delays (Minutes)')
g.set_title('Delayed Flights by origin-destination (USA, 2008)')

plt.show()
```
